```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import tifffile
from tifffile import imread
from matplotlib.backends.backend_pdf import PdfPages
import glob, os, re

from aicsimageio import AICSImage, imread, writers
from skimage import color, io
```

```python
predict_0 = AICSImage('predictions/tifs/0_prediction_c0.model.p.tif')
P_0 = predict_0.get_image_data()
P_0.shape
#np.delete(arr2, 0, 0)

signal_0 = AICSImage('predictions/tifs/0_signal.tif')
S_0 =signal_0.get_image_data()
S_0.shape
#np.delete(arr2, 0, 0)

target_0 = AICSImage('predictions/tifs/0_target.tif')
T_0 = target_0.get_image_data()
T_0.shape
#np.delete(arr2, 0, 0)
```

```python
P_image = np.squeeze(np.squeeze(np.squeeze(P_0, axis=0),axis=1),axis=0)
P_image.shape
#plt.imshow(im_image)

S_image = np.squeeze(np.squeeze(np.squeeze(S_0, axis=0),axis=1),axis=0)
S_image.shape
#plt.imshow(im_image)

T_image = np.squeeze(np.squeeze(np.squeeze(T_0, axis=0),axis=1),axis=0)
T_image.shape
#plt.imshow(im_image)
```

```python
for i in range(0, 1):
  
  fig, axes = plt.subplots(1, 3)
  fig.set_size_inches(15,8 )
  # Plot sample and reconstruciton
  axes[0].imshow(np.max(P_image,axis=0), cmap='gray')
  axes[0].set_title('predicted_image')
  axes[1].imshow(np.max(S_image,axis=0), cmap='gray')
  axes[1].set_title('Signal')
  axes[2].imshow(np.max(T_image,axis=0), cmap='gray')
  axes[2].set_title('Target')
 
  plt.show()
```
